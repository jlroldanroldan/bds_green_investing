```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_colwidth', 100)
```

# Import and Clean Data

```python
sentiment = pd.read_csv('../../../data/team/NLTK Time Series/sentiment_nltk_fixed_final.csv')
```

```python
sentiment.drop(columns='Unnamed: 0', inplace=True)
```

```python
sentiment.dropna(inplace=True)
```

# Create Year-Week column

```python
sentiment['Date'] = pd.to_datetime(sentiment['CreatedAt'].str[:10], errors='coerce')
sentiment['Year-Week'] = sentiment['Date'].dt.strftime('%Y-%U')
```

# Create Weekly mean of Sentiment Scores (neg, pos, compound)

```python
weekly_mean_sentiment.columns
```

```python
weekly_mean_sentiment = sentiment.groupby(['company', 'Year-Week']).mean()
```

```python
weekly_mean_sentiment.reset_index(inplace=True)
```

```python
weekly_mean_sentiment
```

```python
weekly_mean_sentiment.set_index('Year-Week', inplace=True)
```

```python
weekly_mean_sentiment_compound = pd.DataFrame(weekly_mean_sentiment[['company','compound']])
```

```python
weekly_mean_sentiment_compound
```

```python
weekly_mean_compound_sentiment_by_company = weekly_mean_sentiment.pivot( columns='company', values='compound')
```

```python
weekly_mean_compound_sentiment_by_company.drop('tesla',axis=1, inplace=True)
```
