# Import Sentiment 

```python
sentiment = pd.read_csv('../../data/team/NLTK Time Series/sentiment_nltk_fixed_final.csv')
```

```python
sentiment.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
sentiment['Date'] = pd.to_datetime(sentiment['CreatedAt'].str[:10], errors='coerce')
```

```python
sentiment.set_index('Date',inplace=True);
```

```python
sentiment.drop('CreatedAt', axis=1, inplace=True)
```

```python
sentiment
```

## Create daily tweet count 

```python
spwr_tweets = sentiment[sentiment['company'] == 'first_solar']
```

```python
swpr_tweets_daily = spwr_tweets.groupby(['Date']).count()
```

```python
swpr_tweets_daily = pd.DataFrame(swpr_tweets_daily['compound'])
```

```python
swpr_tweets_daily.rename(columns={'compound':'Daily Tweets'},inplace=True)
```

```python
swpr_tweets_daily
```

## Create compount sentiment mean

```python
spwr_sentiment = sentiment[sentiment['company'] == 'first_solar']
```

```python
spwr_sentiment = pd.DataFrame(spwr_sentiment['compound'])
```

```python
spwr_sentiment_daily = spwr_sentiment.groupby(['Date']).mean()
```

```python
spwr_sentiment_daily
```

# Merge SPWR datasets

```python
spwr = pd.merge(spwr_closing_daily, swpr_tweets_daily,how='left', left_on=['Date'], right_on=['Date'] )
```

```python
spwr.dropna(inplace=True)
```

```python
spwr = pd.merge(spwr, spwr_sentiment_daily,how='left', left_on=['Date'], right_on=['Date'] )
```

```python
spwr
```

```python
spwr.rename(columns={'close':'Closing Price', 'compound':'Sentiment Score'}, inplace=True)
```

```python
spwr
```

# Normalized SPWR data

```python
spwr_normalized = spwr.copy()
```

```python
spwr_normalized
```

```python
def normalize_daily_tweets(df):
    max_count = df['Daily Tweets'].max()
    min_count = df['Daily Tweets'].min()
    df['Daily Tweets'] = (df['Daily Tweets'] - min_count) / (max_count - min_count)
    return df

def normalize_sentiment_score(df):
    max_count = df['Sentiment Score'].max()
    min_count = df['Sentiment Score'].min()
    df['Sentiment Score'] = (df['Sentiment Score'] - min_count) / (max_count - min_count)
    return df

def normalize_closing_price(df):
    max_count = df['Closing Price'].max()
    min_count = df['Closing Price'].min()
    df['Closing Price'] = (df['Closing Price'] - min_count) / (max_count - min_count)
    return df
```

```python
spwr_normalized = normalize_daily_tweets(spwr_normalized)
```

```python
spwr_normalized = normalize_sentiment_score(spwr_normalized)
```

```python
spwr_normalized =  normalize_closing_price(spwr_normalized)
```
